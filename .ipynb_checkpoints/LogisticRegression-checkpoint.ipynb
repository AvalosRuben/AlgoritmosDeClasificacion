```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df = pd.read_csv('train.csv')
#Filtrar las variables que s√≠ usaremos del DataFrame
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
X = df[features]
y = df["Survived"]
```

```python
X = X.copy()
age_medians = X.groupby(["Sex", "Pclass"])["Age"].median()
def fill_age(row):
    if pd.isnull(row["Age"]):
        return age_medians.loc[row["Sex"], row["Pclass"]]
    return row["Age"]

X["Age"] = X.apply(fill_age, axis=1)
```

```python
X = X.copy()
fare_medians = X.groupby(["Pclass"])["Fare"].median()
def fill_fare(row):
    if pd.isnull(row["Fare"]):
        return fare_medians.loc[row["Pclass"]]
    return row["Fare"]

X["Fare"] = X.apply(fill_fare, axis=1)
```

```python
embarked_moda = X["Embarked"].mode()[0]
X["Embarked"] = X["Embarked"].fillna(embarked_moda)
```

```python
#Label encoding sex
sex_map = {"male":0, "female":1}

X["Sex"] = X["Sex"].map(sex_map)

X = X.dropna(subset=["Sex"])
X["Sex"] = X["Sex"].astype(int)
```

```python
#One hot encoding embarked
X = pd.get_dummies(X, columns=["Embarked"], drop_first=True)
```

```python
scaler = RobustScaler()

X[["Age", "Fare"]] = scaler.fit_transform(X[["Age", "Fare"]])
```
